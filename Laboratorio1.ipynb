7, Limpiar y preparar el dataset Clientes_Colombia.xlsx para su uso en análisis de
datos. Este laboratorio te ayudará a aplicar diversas técnicas de limpieza de datos
con pandas, incluyendo la gestión de valores faltantes, la eliminación de
duplicados, la normalización de formatos, el manejo de valores atípicos y la
conversión de tipos de datos. Este laboratorio te permitirá desarrollar habilidades
prácticas en la limpieza de datos utilizando pandas. Asegúrate de seguir cada
paso cuidadosamente y justificar cada decisión que tomes durante el proceso. Un
conjunto de datos limpio es fundamental para garantizar la precisión y eficacia de
cualquier análisis posterior

```python
## leemos el archivo externo
Datos_clientes = pd.read_excel("/content/Clientes_Colombia.xlsx")
Datos_clientes
```

```python
Datos_clientes.info ()  ## revisamos el tipo de informacion
```

```python
Datos_clientes.isnull().sum() ##identificamos datos faltantes
```

```python
Datos_clientes["Nombre"] = Datos_clientes["Nombre"].str.strip().str.title()### Limpiamos la columna de  Nombre y apellidos quitando espacios y organizando
Datos_clientes["Apellido"] = Datos_clientes["Apellido"].str.strip().str.title()
Datos_clientes.head(4) ## observamos las primeras 4 columnas para vizualizar el df
```

```python
Datos_clientes['ValorDeCompra (mil)'] = Datos_clientes['ValorDeCompra (mil)'].str.replace(r'[^\d]', '', regex=True) ##Limpiamo la columna retirando el moneda y simbolo
Datos_clientes['ValorDeCompra (mil)'] = pd.to_numeric(Datos_clientes['ValorDeCompra (mil)'], errors='coerce').fillna(0).astype(int)  ##Convertir a enteros. Primero a float para manejar posibles NaN
Datos_clientes.head(4)
```

```python
## Reliazamos un trabajo de limpieza sobre nuestro df
Datos_clientes['Fechas_Diversas'] = pd.to_datetime(Datos_clientes['Fechas_Diversas'], errors='coerce', dayfirst=True) ### Convertir  columnas fechas y fechas diversas  a formato de fecha
Datos_clientes['Fecha'] = pd.to_datetime(Datos_clientes['Fecha'], dayfirst=True)
Datos_clientes['Fecha'] = Datos_clientes['Fecha'].interpolate() ### Interpolacion de datos para agregar infomraicon en datos vacios
Datos_clientes['Fechas_Diversas'] = Datos_clientes['Fechas_Diversas'].fillna(Datos_clientes['Fecha']) ### Reemplazamos los Datos que salen en NAN de la columna Fechas diversas con el valoñr de la columna Fechas
Datos_clientes['Fechas_Diversas'] = Datos_clientes['Fechas_Diversas'].dt.strftime('%Y-%m-%d') # estanadarizamo la columan fecha diversas
Datos_clientes['Fechas_Diversas'].fillna('Valor_Predeterminado', inplace=True)
Datos_clientes['Nombre'].replace('', 'desconocido', inplace=True) ## llenamos los espacion vacion de la colimna nombre con la palabra Desconocidos
Datos_clientes['Nombre'].fillna('desconocido', inplace=True) ## eveluamos tambien los datos nulos
Datos_clientes['Apellido'].fillna('desconocido', inplace=True) ## llenamos los espacion vacion de la colimna Apellido con la palabra Desconocidos
Datos_clientes = Datos_clientes.drop_duplicates() ## Eliminamos los duplicados
Datos_clientes
```
